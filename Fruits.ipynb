```
get_ipython().magic('matplotlib inline')
import cv2
import matplotlib
from matplotlib import colors
from matplotlib import pyplot as plt
import numpy as np
```

FRUIT 1 - *STRAWBERRY* 


```
!wget "https://image.shutterstock.com/z/stock-photo-ripe-and-unripe-strawberry-50413600.jpg" -O dt.jpg
image = cv2.imread('./dt.jpg')
plt.imshow(image)
plt.show()
#imshow reads color in BGR(blue,green,red)
image.shape
```

```
# Convert from BGR to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Resize to a third of the size
image = cv2.resize(image, None, fx=1/3, fy=1/3)
plt.figure(figsize=(15, 15))
    
    # Show image, with nearest neighbour interpolation
plt.imshow(image, interpolation='nearest')
```

```
def display_RGB(image):
    row,col,plane = image.shape
    temp = np.zeros((row,col,plane),np.uint8)
    temp[:,:,0] = image[:,:,0]

  # displaying the Red plane image
    plt.imshow(temp)
    plt.show()

  # again take a zero matrix of image matrix shape
    temp = np.zeros((row,col,plane),np.uint8)
    temp[:,:,1] = image[:,:,1]

  # displaying the Green plane image
    plt.imshow(temp)
    plt.show()


  # again take a zero matrix of image matrix shape
    temp = np.zeros((row,col,plane),np.uint8)
    temp[:,:,2] = image[:,:,2]

  # displaying the Blue plane image
    plt.imshow(temp)
    plt.show()
```

```
display_RGB(image)
```

```
# Convert from RGB to HSV
hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
plt.imshow(hsv)
images = []
for i in [0, 1, 2]:
    colour = hsv.copy()
    if i != 0: colour[:,:,0] = 0
    if i != 1: colour[:,:,1] = 255
    if i != 2: colour[:,:,2] = 255
    images.append(colour)
hsv_stack = np.vstack(images)
rgb_stack = cv2.cvtColor(hsv_stack, cv2.COLOR_HSV2RGB)
plt.figure(figsize=(15, 15))
# Show image, with nearest neighbour interpolation
plt.imshow(rgb_stack, interpolation='nearest')
    
```

```
# Blur image slightly
image_blur = cv2.GaussianBlur(image, (7, 7), 0)
plt.figure(figsize=(15, 15))
# Show image, with nearest neighbour interpolation
plt.imshow(image_blur, interpolation='nearest')
```

```
image_blur_hsv = cv2.cvtColor(image_blur, cv2.COLOR_RGB2HSV)
# 170-180 hue
min_red2 = np.array([170, 100, 60])
max_red2 = np.array([180, 256, 256])
image_red2 = cv2.inRange(image_blur_hsv, min_red2, max_red2)
plt.figure(figsize=(10, 10))
plt.imshow(image_red2,cmap='gray')
```

```
# to get elliptical shaped kernels we use the function cv2.getStructuringElement()
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (15, 15))

#to close small holes
image_red_closed = cv2.morphologyEx(image_red2, cv2.MORPH_CLOSE, kernel)
plt.figure(figsize=(10, 10))
plt.imshow(image_red_closed,cmap='gray')


```

```
# Convert from BGR to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Resize to a third of the size
image = cv2.resize(image, None, fx=1/3, fy=1/3)
plt.figure(figsize=(15, 15))
    
    # Show image, with nearest neighbour interpolation
plt.imshow(image, interpolation='nearest')
```

```
display_RGB(image)
```

```
# Convert from RGB to HSV
hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
plt.imshow(hsv)
images = []
for i in [0, 1, 2]:
    colour = hsv.copy()
    if i != 0: colour[:,:,0] = 0
    if i != 1: colour[:,:,1] = 255
    if i != 2: colour[:,:,2] = 255
    images.append(colour)
hsv_stack = np.vstack(images)
rgb_stack = cv2.cvtColor(hsv_stack, cv2.COLOR_HSV2RGB)
plt.figure(figsize=(15, 15))
# Show image, with nearest neighbour interpolation
plt.imshow(rgb_stack, interpolation='nearest')
```

```
image_blur_hsv = cv2.cvtColor(image_blur, cv2.COLOR_RGB2HSV)
yellow_lower = np.array([20, 100, 100])
yellow_upper = np.array([30, 255, 255])
mask_yellow = cv2.inRange(image_blur_hsv, yellow_lower, yellow_upper)
plt.figure(figsize=(10, 10))
plt.imshow(mask_yellow,cmap='gray')
```

```
# to get elliptical shaped kernels we use the function cv2.getStructuringElement()
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (15, 15))

#closes small holes
image_yellow_closed = cv2.morphologyEx(mask_yellow, cv2.MORPH_CLOSE, kernel)
plt.figure(figsize=(10, 10))
plt.imshow(image_yellow_closed,cmap='gray')
```
